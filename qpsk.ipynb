# Signal DownConersion
This will take up to 7 steps which are
- Read time series data from csv or hdf5 files
- Plot the data read for the aim of showing how noisy the signal is
- Signal Filtering. The carrier is at 2GHz with data at  GHz (in phase) and 3GHz(quadrature) the bandwidth of the signal is 2GHz. Low pass filter and band pass filter are going to be used
- Down conversion. At this step carrier signal at 2GHz needs to be removed. This will have us remain with data at 1Ghz and 3GHz
- low pass filter at 1.5Ghz bandwidth
- Time recovery
- Phase (Carrier) recovery

> The last two steps will enable us to plot eye and constelattion diagram of the in-phase and quadrature

- Analysis of the in-phase and quadrature using parameter such as BER( Bit Error Rate) and EVM

### 1. reading time series data from csv file

````python
from pandas import read_csv
series = read_csv('/home/zeus/Desktop/qpsk/data/1Gbps_Data_20Gbps.csv',header=None, delimiter=',', names=["time","data"])

#getting all the amplitude data
time = series["time"]
amp = series["data"]
print(time)
print(amp)
````

### 2. ploting the data

> The purpose of this step is to show how noisy the data is: For the sole purpose of clear visibility the dataset is not plotted wholy, only 5000 points are plotted

The code
```python
plt.plot(amp[:5000])
```
determines how many data points from 0 should be plotted. To plot the whole dataset the line should be changed to

```python
plt.plot(amp)
```

````python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (20,15)
plt.plot(amp[:5000])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
````
